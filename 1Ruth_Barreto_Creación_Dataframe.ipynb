# **Descripción de las variables**
* Latitud y longitud son de tipo geográfico.
* Región y departamento son  ubicación geográfica.
* Temperatura es de tipo térmica.
* Velocidad del Viento y Dirección del Viento	 son variables de tipo Dirección del desplazaiento.
* Presión es el peso de la masa de aire sobre la tierra.
* Punto de rocio es la temperatura a la que empiesa a condensarse el agua presente en la atmosfera.
* La cobertura nubosa es la porción de cielo cubierto por nubes.
* La precipitación es la cantidad de agua que cae desde la atmósfera tras presentarse lluvias.
* Probabilidad de Tormenta es la probabilidad de que se presenten tormentas.
* La humedad es la cantidad de vapor de agua que hay en la atmósfera.
* El pronóstico es la predicción de como estara el cielo.

#  **Llenamos los datos Faltantes**

* En este caso los datos faltantes corresponden a la **Región** el cual no es un atributo de utilidad y del cual se puede prescindir, por lo que eliminamos dicha columna de la base de datos.

```python
caucasia = caucasia.dropna(axis=1)
caucasia
```

# **Codificación de las Variables Categóricas**
* Para codificar las variables categoricas es necesario identificar si esta es ordinal o nominal esto con el fin de utilizar un método que nos de una codificación correcta.
* Continuación tenemos una variable categorica nominal, es decir, no hay un orden, por lo que: 
*Utilizaremos **sklearn** para la codificación y **OneHot** para la representación gráfica, la cual nos entregara una matriz donde se daran valores de 0 y 1 , donde **0 es no** y **1 sí** para las caracteristicas definidas. 

```python
caucasia_cat = caucasia[['Pronóstico ']] 
```

```python
from sklearn.preprocessing import OneHotEncoder  # con sklearn codificamos variables catagóricas nominales
codificador_onehot = OneHotEncoder() # con OneHot hacemos la representación vectorial
caucasia_cat_codificado_onehot = codificador_onehot.fit_transform(caucasia_cat)
a = caucasia_cat_codificado_onehot.toarray() 
codificador_onehot.categories_
caucasia_cat_codificado_onehot_df = pd.DataFrame(a, columns =  ['Parcialmente Nublado', 'Nublado', 'Nublado - Llovizna'])
caucasia_cat_codificado_onehot_df
```
